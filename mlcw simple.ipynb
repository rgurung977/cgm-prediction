**First creating a new folder for storing the renamed csv files**

```python
import os

# Set the folder path
folder_path = "MAL2324_CW_DataSet_Initial"
folder_path2 = "mlcw2"
# Get list of all files in the folder
files = os.listdir(folder_path)

# Loop through each file
for filename in files:
    # Check if the file has '#' in its name
    if '#' in filename:
       
        # Create new filename by replacing '#' with nothing
        new_filename = filename.replace('#', '')
        
        # Get full paths for old and new names
        old_path = os.path.join(folder_path, filename)
        new_path = os.path.join(folder_path2, new_filename)
        
        # Rename the file
        os.rename(old_path, new_path)
        print(f"Renamed: {filename} -> {new_filename}")

print("All files have been renamed!")
```

```python


# Create a label column (0 = normal, 1 = hypo, 2 = hyper)
labels = []

CGM = adolescent001["CGM"].values
look_ahead = 24  # 2 hours = 24 samples

for i in range(len(CGM)):
    future_window = CGM[i+1 : i+1+look_ahead]

    if len(future_window) < look_ahead:
        labels.append(None)  # not enough future data
        continue

    if (future_window < 70).any():
        labels.append(1)   # hypo
    elif (future_window > 180).any():
        labels.append(2)   # hyper
    else:
        labels.append(0)   # normal

adolescent001["label"] = labels

print(adolescent001.shape)
```

```python
# Create a label column (1 = hypo, 1 = hyper)
hypocgm = []
hypercgm = []
normal = []

CGM = adolescent001["CGM"].values
look_ahead = 24  # 2 hours = 24 samples

for i in range(len(CGM)):
    future_window = CGM[i+1 : i+1+look_ahead]

    if len(future_window) < look_ahead:
        hypocgm.append(0)  # not enough future data
        hypercgm.append(0)
        normal.append(1)
        continue

    if (future_window < 70).any():
        hypocgm.append(1)   # hypo
        hypercgm.append(0)
        normal.append(0)
    elif (future_window > 180).any():
        hypocgm.append(0)   # hyper
        hypercgm.append(1)
        normal.append(0)
    else:
        hypocgm.append(0)   # normal
        hypercgm.append(0)
        normal.append(1)

ado1["hypocgm"] = hypocgm
ado1["hypercgm"] = hypercgm
ado1["normal"] = normal

print(ado1.shape)
```

***To visualize the number of normal cases vs abnormal cases***

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# using seaborn countplot

plt.figure(figsize=(8, 6))
sns.countplot(data=ado1, 
              x = 'normal')
plt.title('Distribution of normal events', fontsize=14)
plt.xlabel('normal(0 = event, 1 = normal)', fontsize= 12)
plt.ylabel('count', fontsize=12)

# adding labels on the bars

for i, v in enumerate(ado1['normal'].value_counts().values):
    plt.text(i, v+50, str(v), ha = 'center', fontsize=11)

plt.show()
```
